```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
from scipy.interpolate import interp1d
from scipy import stats

def rescale_ts(ts, length=100):
    x = np.linspace(0, len(ts)-1, num=len(ts))
    y = np.array(ts)

    f = interp1d(x, y)
    new_x = np.linspace(0, len(ts)-1, num=length)
    interpolated_y = f(new_x)
    
    return stats.zscore(interpolated_y)
```

```python
data_class = 'Adiac'
f_name = f'ClassificationClusteringDatasets/{data_class}_TEST'

with open(f_name, 'r') as f:
    content = f.read()
    ts = []
    for line in content.split('\n'):
        if len(line.split()) == 0:
            continue
        ts.append([float(i) for i in line.split()])
```

```python
for data in ts:
    data = rescale_ts(data)
    plt.plot(data, alpha=0.25)
plt.title('Adiac')
plt.show()
```
